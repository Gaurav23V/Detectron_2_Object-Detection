```python
import os

# Set the base directory
base_dir = "/content"

# Create the main 'data' directory
data_dir = os.path.join(base_dir, "data")
os.makedirs(data_dir, exist_ok=True)

# Create subdirectories for train, val, and test
train_dir = os.path.join(data_dir, "train")
val_dir = os.path.join(data_dir, "val")
test_dir = os.path.join(data_dir, "test")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Create subdirectories for imgs and anns
train_imgs_dir = os.path.join(train_dir, "imgs")
train_anns_dir = os.path.join(train_dir, "anns")
val_imgs_dir = os.path.join(val_dir, "imgs")
val_anns_dir = os.path.join(val_dir, "anns")
test_imgs_dir = os.path.join(test_dir, "imgs")
test_anns_dir = os.path.join(test_dir, "anns")

os.makedirs(train_imgs_dir, exist_ok=True)
os.makedirs(train_anns_dir, exist_ok=True)
os.makedirs(val_imgs_dir, exist_ok=True)
os.makedirs(val_anns_dir, exist_ok=True)
os.makedirs(test_imgs_dir, exist_ok=True)
os.makedirs(test_anns_dir, exist_ok=True)

print("Directory structure created successfully!")
```
